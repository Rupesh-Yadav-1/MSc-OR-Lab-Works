```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def G_h(z):
  if 1-z > 0:
    return 1-z
  else:
    return 0
```

```python
def G_l(z):
  return np.log(1+np.exp(-z))
```

```python
def G_sh(z):
  if 1-z > 0:
    return (1-z)**2
  else:
    return 0
```

```python
z = np.linspace(-100, 100, 1000)
plt.figure(figsize=(10,9))
plt.plot(z,[G_h(i) for i in z], label = 'hinge')
plt.plot(z,[G_l(i) for i in z], label = 'logistic')
plt.plot(z,[G_sh(i) for i in z], label = 'sq. hinge')
plt.legend()
plt.show()
```
