```python
import math
x1=2.34
x2=-5.67
y=1.5

print('Finding e^x :')
print('for x= 2.34 : ',math.exp(x1))
print('for x= -5.67 : ',math.exp(x2))

print('Finding |x| :')
print('for x= 2.34 : ',math.fabs(x1))
print('for x= -5.67 : ',math.fabs(x2))

print('Finding ln|x| :')
print('for x= 2.34 : ',math.log(x1))
print('for x= -5.67 : ',math.log(math.fabs(x2)))

print('Finding |x|^y (where y=1.5):')
print('for x= 2.34 : ',math.pow(x1,y))
print('for x= -5.67 : ',math.pow(math.fabs(x2),y))

print('Integer part of x1=2.34 : ',math.floor(x1))
print('Integer part of x2=-5.67 : ',math.floor(x2))
print('Integer part of y=1.5 : ',math.floor(y))
print('fractional part of x1=2.34 : ',x1-math.floor(x1))
print('fractional part of x2=-5.67 : ',x2-math.floor(x2))
print('fractional part of y=1.5 : ',y-math.floor(y))

print('cos(\u03C0): ',math.cos(math.pi))
print('cos(\u03C0/4): ',math.tan(math.pi/4))

print('\u03C0/4 in degrees is : ',math.degrees(math.pi/4))
```

```python
import matplotlib.pyplot as plt
x=np.linspace(0,2,200)
f=np.exp(-x)*np.cos(6*np.pi*x)
g=np.exp(-x)
h=-np.exp(-x)
plt.plot(x,f,label=r'$f(x)=\ exp(-x)cos(6pix)$')
plt.plot(x,g,label=r'$g(x)=\ exp(-x)$')
plt.plot(x,h,label=r'$h(x)=\ -exp(-x)$')
plt.title('Graphs of given three functions')
plt.xlabel(r'$x$',fontsize=14)
plt.ylabel(r'Function of $x$',fontsize=14)
plt.legend(loc='lower right')
plt.show()
```

```python
from numpy import random
mu,l = (0.6,10)
x=random.poisson(mu,l)
print('10 random values which follows the Poisson distribution(parameter, mu=0.6) are: ')
print(x)
print()
print('The CDF of each value is as follows: ')
from scipy.stats import poisson
for num in x:
  x,mu = (num,0.6)
  F_x=poisson.cdf(x,mu)
  print('F(',x,') = ',F_x)

print()

lm,l = (0.6,10)
y=random.exponential(lm,l)
print('10 random values which follows the Exponential distribution(parameter, lambda=0.6) are: ')
print(y)
print()
print('The CDF of each value is as follows: ')
from scipy.stats import expon
for no in y:
  y,lm = (no,0.6)
  G_y=expon.cdf(y,lm)
  print('F(',y,') = ',G_y)
```
