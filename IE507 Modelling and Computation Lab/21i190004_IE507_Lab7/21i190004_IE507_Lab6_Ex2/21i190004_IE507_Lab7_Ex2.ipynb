```python
import matplotlib.pyplot as plt
plot=sns.barplot(x="mpg",y="displacement",data=df)

x=plot.patches[3].get_x() + plot.patches[3].get_width() / 2
y=plot.lines[3].get_ydata()[1]+0.3
plt.annotate('highest bar', xy=(x,y),xytext=(x+0.3,y+55), arrowprops=dict(facecolor='red'))
```

**Part 4**

```python
import numpy as np
```

```python
pl=sns.barplot(x="mpg",y="displacement",hue='model_year',data=df,estimator=np.median)
pl.legend(loc='upper right',ncol=3)
```

Since there are lot of values for mpg hence we are getting large numbers of bars.

Those cars which have low mpg are likely to have more displacement value.

```python
plot = sns.barplot(x="mpg",y="displacement",hue="model_year",data=df,estimator=np.median)
for p,line in zip(plot.patches, plot.lines):
    plot.annotate(format(p.get_height(),'.2f'), 
                   (p.get_x() + p.get_width() / 2, line.get_ydata()[1]), 
                   ha = 'center', va = 'top', 
                   xytext = (0, 9), 
                   textcoords = 'offset points')

plot.legend(loc='upper right',ncol=4)
```

```python
k=0
for i in range(len(df['displacement'])):
  if df['displacement'][i]==68:
    k=k+1
    j=i

print(k,j)
#means in 118th row displacement is minimum(68)
#implies bar corresponding to mpg=df['mpg'][117] is lowest
```
