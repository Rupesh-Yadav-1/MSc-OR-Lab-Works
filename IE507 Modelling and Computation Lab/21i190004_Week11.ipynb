```python
f = open("FlowShop1.txt", "r")
n_job=int(f.readline().rstrip())
f.close()
```

```python
data = pd.read_csv('FlowShop1.txt')
data.to_csv('df.csv')
df = pd.read_csv('df.csv')
```

```python
listA=list(df['Unnamed: 0'])
listB=list(df[str(n_job)])
```

```python
def fifo():
  list1=[]  #list for completion time of jobs on Machine A
  for i in range(n_job):
    j=0
    sum_i=0
    while j<=i:
      sum_i=sum_i+listA[j]
      j=j+1
    list1.append(sum_i)
  list2=[]     #list for completion time of jobs on Machine B
  list2.append(list1[0]+listB[0])
  k=1
  while k<n_job:
    if list1[k]>=list2[k-1]:
      list2.append(list1[k]+listB[k])
    else:
      list2.append(list2[k-1]+listB[k])
    k=k+1
  
  print('Job sequence:')
  for i in range(len(list2)):
    print(i+1,',',end='')
  print()
  print()
  print('Completion time:')
  for i in range(len(list2)):
    print(str(i+1)+':'+str(list2[i]),',',end='')
  print()
  print()
  print('Makespan:')
  print(list2[n_job-1])

  return 
```

```python
def john():
  listA1=listA.copy()
  listB1=listB.copy()
  seq_1=[]
  seq_2=[]
  p=0
  while p < n_job:
    if min(listA1) <= min(listB1):
      ind1=listA1.index(min(listA1))
      #ind2=listA.index(min(listA1))
      seq_1.append(ind1+1)
      listA1[ind1]=max(listA)+max(listB)+100
      listB1[ind1]=max(listA)+max(listB)+100
    else:
      ind1=listB1.index(min(listB1))
      #ind2=listB.index(min(listB1))
      seq_2.append(ind1+1)
      listA1[ind1]=max(listA)+max(listB)+100
      listB1[ind1]=max(listA)+max(listB)+100
    p=p+1
  #print(seq_2)
  seq_2.reverse()
  seq=seq_1+seq_2

  list1=[]  #list for completion time of jobs on Machine A
  for i in range(len(seq)):
    j=0
    sum_i=0
    while j<=i:
      sum_i=sum_i+listA[seq[j]-1]
      j=j+1
    list1.append(sum_i)
  list2=[]     #list for completion time of jobs on Machine B
  list2.append(list1[0]+listB[seq[0]-1])
  k=1
  while k<n_job:
    if list1[k]>=list2[k-1]:
      list2.append(list1[k]+listB[seq[k]-1])
    else:
      list2.append(list2[k-1]+listB[seq[k]-1])
    k=k+1

  print('Job sequence:')
  print(seq)
  print()
  print('Completion time:')
  for i in range(len(list2)):
    print(str(i+1)+':'+str(list2[i]),',',end='')
  print()
  print()
  print('Makespan:')
  print(list2[n_job-1])

  return
```

```python
def stpt():
  sq=[a + b for a, b in zip(listA, listB)]
  sq2=sq.copy()
  sq.sort()
  seq=[]
  for i in sq:
    seq.append(sq2.index(i)+1)
    sq2[sq2.index(i)]=max(sq)+10
  
  list1=[]  #list for completion time of jobs on Machine A
  for i in range(len(seq)):
    j=0
    sum_i=0
    while j<=i:
      sum_i=sum_i+listA[seq[j]-1]
      j=j+1
    list1.append(sum_i)
  list2=[]     #list for completion time of jobs on Machine B
  list2.append(list1[0]+listB[seq[0]-1])
  k=1
  while k<n_job:
    if list1[k]>=list2[k-1]:
      list2.append(list1[k]+listB[seq[k]-1])
    else:
      list2.append(list2[k-1]+listB[seq[k]-1])
    k=k+1

  print('Job sequence:')
  print(seq)
  print()
  print('Completion time:')
  for i in range(len(list2)):
    print(str(i+1)+':'+str(list2[i]),',',end='')
  print()
  print()
  print('Makespan:')
  print(list2[n_job-1])

  return
```

```python
n=int(input('Enter 1 for FIFO, 2 for Johnson Rule, 3 for Shortest Total Processing time: '))
if n==1:
  print('According to FIFO:')
  print()
  fifo()
elif n==2:
  print('According to johnson rule:')
  print()
  john()
elif n==3:
  print('According to shortest total processing time:')
  print()
  stpt()
else:
  print('invalid input!')
```
