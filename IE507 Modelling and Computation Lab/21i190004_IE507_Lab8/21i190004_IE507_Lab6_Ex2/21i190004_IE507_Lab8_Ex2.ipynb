```python
#importing the required packages
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv('b4.txt', sep='    ',index_col=None,header = None)
```

```python
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
X = df.to_numpy()
```

```python
import matplotlib.pyplot as plt
```

```python
klist=[3,5,7,9,11,13,15,17,19]
```

```python
for j in klist:
  print('For K=',j)
  km = KMeans(n_clusters=j)
  km_result = km.fit(X)
  label = km_result.labels_
  #Getting the Centroids
  centroids = km.cluster_centers_
  u_labels = np.unique(label)
 
  #plotting the results:
  
  for i in u_labels:
    plt.scatter(X[label == i , 0] , X[label == i , 1] , label = i)
  plt.scatter(centroids[:,0] , centroids[:,1] , s = 50, color = 'k')
  plt.legend()
  plt.show()
```

**Part 3**

As K increases clusters becomes more symmetric about their respective centers.

**Part 4**

```python
t = pd.read_csv('test.txt', sep='    ',index_col=None,header = None)
```

```python
test = t.to_numpy() 
```

```python
markers=["p","<","*","h",">","1","2","3"]
```

```python
for i in klist:
  print('For K=',i)
  km = KMeans(n_clusters=i)
  km_result = km.fit(X)
  label = km_result.labels_
  u_labels = np.unique(label)
  print('Predicted labels respectively are as follows: ')
  print(km_result.predict(test))
  
  for i in u_labels:
    plt.scatter(X[label == i , 0] , X[label == i , 1] , label = i)
  for j in range(len(test)):
    plt.scatter(x=test[j,0],y=test[j,1],color='black',marker=markers[j], s=80)
  plt.legend()
  plt.show()
```

**Part 5**

When K was low prediction were better.

**Part 6**

```python
cost=[]
for i in klist:
  KM=KMeans(n_clusters=i)
  KM.fit(X) 
  cost.append(KM.inertia_) #it calculates squared error(cost) corrs. to K
plt.plot(klist,cost)
plt.xlabel('Value of K')
plt.ylabel('squared error(cost)')
plt.show
```

K=5 is the best choice.

**Part 5 and 9**

```python
df_norm = df.copy()
df_norm = df_norm.astype(float)
for i in range(len(df[0])):
  df_norm[0][i] = (df_norm[0][i]-df[0].mean())/df[0].std()

for i in range(len(df[1])):
  df_norm[1][i] = (df_norm[1][i]-df[1].mean())/df[1].std()
```

```python
Y=df_norm.to_numpy()
```

```python
for j in klist:
  print('For K=',j)
  km = KMeans(n_clusters=j)
  km_result = km.fit(Y)
  label = km_result.labels_
  #Getting the Centroids
  centroids = km.cluster_centers_
  u_labels = np.unique(label)
 
  #plotting the results:
  
  for i in u_labels:
    plt.scatter(Y[label == i , 0] , Y[label == i , 1] , label = i)
  plt.scatter(centroids[:,0] , centroids[:,1] , s = 50, color = 'k')
  plt.legend()
  plt.show()
```

**Part 10**

As K increases clusters becomes more round shaped and symmetric about their centers.

**Part 11**

```python
t_norm = t.copy()
t_norm = t_norm.astype(float)
for i in range(len(t[0])):
  t_norm[0][i] = (t_norm[0][i]-df[0].mean())/df[0].std()

for i in range(len(t[1])):
  t_norm[1][i] = (t_norm[1][i]-df[1].mean())/df[1].std()
```

```python
test_norm = t_norm.to_numpy()
```

```python
for i in klist:
  print('For K=',i)
  km = KMeans(n_clusters=i)
  km_result = km.fit(Y)
  label = km_result.labels_
  u_labels = np.unique(label)
  print('Predicted labels respectively are as follows: ')
  print(km_result.predict(test_norm))
  
  for i in u_labels:
    plt.scatter(Y[label == i , 0] , Y[label == i , 1] , label = i)
  for j in range(len(test)):
    plt.scatter(x=test_norm[j,0],y=test_norm[j,1],color='black',marker=markers[j], s=80)
  plt.legend()
  plt.show()
```

**Part 12**

Predictions are more accurate when K was low.

**Part 13**

```python
cost2=[]
for i in klist:
  KM2=KMeans(n_clusters=i)
  KM2.fit(Y) 
  cost2.append(KM2.inertia_) #it calculates squared error(cost) corrs. to K
plt.plot(klist,cost2)
plt.xlabel('Value of K')
plt.ylabel('squared error(cost)')
plt.show
```

**Part 14**
